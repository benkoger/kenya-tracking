```python
import cv2
```

```python
video_file = '/media/golden/72FFC6EE48B5CF39/test_tracking_buffalo_v1.mp4' 
new_video_file = '/media/golden/72FFC6EE48B5CF39/test_tracking_video_buffalo_v1_10x.avi'
 

cap = cv2.VideoCapture(video_file)
if cap is None or not cap.isOpened():
    raise Exception('Opencv unable to open the video file for %s' %video_file)
    

fourcc = cv2.VideoWriter_fourcc(*'ffv1')
out = cv2.VideoWriter(filename=new_video_file, fourcc=fourcc, fps=30, 
                              frameSize=(int(cap.get(3)), int(cap.get(4))), isColor=True)
        
good_frame = True 

frame_num = 0

while frame_num < 3700:
    if frame_num % 10 == 0:
        good_frame, image = cap.read()
        out.write(image)
    if not good_frame:
        print('false', frame_num)
    frame_num += 1
    
    
    if frame_num % 50 == 0:
        print(frame_num)
    
cap.release()
cv2.destroyAllWindows()
```
