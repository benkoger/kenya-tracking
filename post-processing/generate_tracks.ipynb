```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm
plt.style.use('dark_background')
```

```python
number_of_tracks = 50
length_of_track = 1000
step_size_scale = 1
max_starting_x = 200
max_starting_y = 200
step_dispersion = 20
chisquare_freedom = 2

```

```python
positions = []
classes = []

track_heading = np.random.rand(number_of_tracks) * 2 * np.pi
track_heading = np.expand_dims(track_heading, 1)

initial_x_position = np.random.rand(number_of_tracks, 1) * max_starting_x 
initial_y_position = np.random.rand(number_of_tracks, 1) * max_starting_y
initial_position = np.hstack((initial_x_position, initial_y_position))

positions.append(initial_position)

for step in range(length_of_track):
    track_heading = np.random.vonmises(track_heading, step_dispersion, (number_of_tracks, 1))
    step_length = np.random.chisquare(chisquare_freedom, (number_of_tracks, 1)) * step_size_scale
    x_position = np.cos(track_heading)
    y_position = np.sin(track_heading)
    new_position = np.hstack((x_position, y_position))
    new_position = new_position * step_length
    new_position = positions[-1] + new_position
    positions.append(new_position)
    
```

```python
plt.figure()
for step in range(length_of_track):
    plt.scatter(positions[step][:,0], positions[step][:,1], c = np.arange(number_of_tracks), cmap=cm.Dark2)
```

```python
import os
#save tracks
new_tracks_name = '{}_tracks_{}_length'.format(number_of_tracks, length_of_track)
print('The track name ', new_tracks_name)
tracks_folder = '/media/golden/72FFC6EE48B5CF39/kenya-tracking/synthetic_tracks' 
new_tracks_folder = os.path.join(tracks_folder, new_tracks_name)
if not os.path.exists(new_tracks_folder):
    os.mkdir(new_tracks_folder) 
positions_file = os.path.join(new_tracks_folder, 'positions.npy') 
np.save(positions_file, positions)
```
